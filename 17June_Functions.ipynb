# 1- positional argument 

```python
def person(name,age):
    print(name)
    print(age)
    
person('nit')
```

```python
def person(name):
    print(name)
    print(age)
    
person('nit', 23) 
```

```python
def person(name, age):
    print(name)
    print(age)
    
person('nit', 23) 
```

```python
def person(age, name):
    print(name)
    print(age)
    
person('nit', 23) 
```

```python
def person(age, name):
    print(name)
    print(age+1)
    
person('nit', 23)
```

# 2- Keyword Argument 

```python
def person(age, name):
    print(name)
    print(age+1)

person('nit', 23)
```

```python
def person(age, name):
    print(name)
    print(age+1)

person(name = 'nit', age = 23)  
```

```python
def person(age, name, mob):
    print(name)
    print(age+1)
    print(mob)

person(name = 'nit', age = 23, 12345)  
```

```python
def person(age, name, mob):
    print(name)
    print(age+1)
    print(mob)

person(name = 'nit', age = 23, mob = 12345)  
```
