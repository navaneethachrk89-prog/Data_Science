# 1- positional argument 

```python
def person(name,age):
    print(name)
    print(age)
    
person('nit')
```

```python
def person(name):
    print(name)
    print(age)
    
person('nit', 23) 
```

```python
def person(name, age):
    print(name)
    print(age)
    
person('nit', 23) 
```

```python
def person(age, name):
    print(name)
    print(age)
    
person('nit', 23) 
```

```python
def person(age, name):
    print(name)
    print(age+1)
    
person('nit', 23)
```

# 2- Keyword Argument 

```python
def person(age, name):
    print(name)
    print(age+1)

person('nit', 23)
```

```python
def person(age, name):
    print(name)
    print(age+1)

person(name = 'nit', age = 23)  
```

```python
def person(age, name, mob):
    print(name)
    print(age+1)
    print(mob)

person(name = 'nit', age = 23, 12345)  
```

```python
def person(age, name, mob):
    print(name)
    print(age+1)
    print(mob)

person(name = 'nit', age = 23, mob = 12345)  
```

# 3- Default Argument 

```python
def person(name, age):
    print(name)
    print(age)

person('nit')
```

```python
def person(name, age=18):
    print(name)
    print(age)

person('nit')
```

```python
def person(name, age=20):
    print(name)
    print(age)

person('nit', 24)
```

# 4- Variable length Argument 

```python
def sum(a, b):
    c = a+b
    print(c)
    
sum(5,6) 
```

```python
def sum(a, b): 
    c = a+b
    print(c)
    
sum(5,6,7,8)  
```

```python
def sum(a, *b): 
    c = a+b
    print(c)
    
sum(5,6,7,8) 
```

```python
def sum(a, *b): # 1st argument is fixed but for 2nd argument
  
    print(type(a))
    print(type(b))
    
    print(a)
    print(b)

sum(5,6,7,8)
```

```python
def sum(a, *b): # 1st argument is fixed & we fetch each value from the tuple & we can add them. 
    c = a   
    
    for i in b:
        c = c + i 
    print(c)
        
sum(5,6,7,8) 
```

```python
def sum(a, *b): # 1st argument is fixed & we fetch each value from the tuple & we can add them. 
    c = a   
    
    for i in b:
        c = c + i 
    print(c)
        
sum(5,6,7,8, 9, 10, -1, -2, 100) 
```

```python
# KWARGS  ( KW + ARGS )

def person(name,*data):
    print(name)
    print(data)

person('ALEX', 36, 'JOHN', 987767)
```

```python
def person(name,*data):
    print(name)
    print(data)

person('ALEX', 36, 'JOHN', 987767)
```

```python
def person(name, *data):
    print('name')
    print(data)

person('ALEX', age = 36, home_place ='southcity', mob =987767)
```

```python
def person(name, **data):
    print('name')
    print(data)

person('ALEX', age = 36, home_place ='southcity', mob =987767)
```
